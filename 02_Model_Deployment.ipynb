```python
# Model Deployment
import pandas as pd
import pickle as p
load_model = p.load(open("Finalized_model.sav",'rb'))

# Default input for ("R&D Spend", "Administration", "Marketing Spend", "State_California", "State_Florida", "State_New York")
x = 245323.53    #R&D Spend
y = 355634.74    #Administration
z = 754242.73    #Marketing Spend
m = 0            #State_California
n = 1            #State_Florida
o = 0            #State_New York

a=pd.DataFrame([[x,y,z,m,n,o]],columns=["R&D Spend", "Administration", "Marketing Spend", "State_California", "State_Florida", "State_New York"])
prediction = load_model.predict(a)
result = prediction.item()
print(f"The Predicted Output is {result:.2f}")
```
